# Setting up transmorph within a scanpy workflow

```python
import scanpy as sc
import warnings

sc.settings.verbosity = 1
sc.settings.set_figure_params(dpi=80, frameon=False, figsize=(3, 3), facecolor='white')
warnings.filterwarnings('ignore')
```

```python
adata_pbmc3k = sc.datasets.pbmc3k_processed()
adata_pbmc68k = sc.datasets.pbmc68k_reduced()

# We create a container for easier handling
adatas = {
    "pbmc3k": adata_pbmc3k,
    "pbmc68k": adata_pbmc68k
}
```

```python
sc.pl.umap(adatas['pbmc3k'], color=['louvain'])
```

```python
sc.pl.umap(adatas['pbmc68k'], color=['louvain'])
```

```python
from transmorph.models import EmbedMNN

embedmnn_model = EmbedMNN()
embedmnn_model.transform(adatas)
```

```python
adatas['pbmc3k'].uns['neighbors']
```

```python
sc.pl.scatter(adatas['pbmc3k'], basis="transmorph")
```
