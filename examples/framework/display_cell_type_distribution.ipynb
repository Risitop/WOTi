```python
from transmorph.utils.plotting import plot_label_distribution_heatmap
from matplotlib.pyplot import rc_context
```

```python
from transmorph.datasets import load_zhou_10x

zhou_10x = load_zhou_10x()
list_anndatas = list(zhou_10x.values())

with rc_context({'figure.figsize': (5, 6)}):
    plot_label_distribution_heatmap(list_anndatas, "class_type", zhou_10x.keys())
```

```python
from transmorph.datasets import load_travaglini_10x

trav_10x = load_travaglini_10x()
list_anndatas = list(trav_10x.values())

with rc_context({'figure.figsize': (1, 1)}):
    plot_label_distribution_heatmap(list_anndatas, "class", trav_10x.keys())
```
